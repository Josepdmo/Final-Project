```python
import pandas as pd

# Supongamos que ya tienes cargada la tabla original en el DataFrame 'book_details_df'

# 1. Seleccionar las columnas necesarias para la tabla Books
books_df = df[['book_id', 'book_title', 'book_details', 'publication_info', 'num_ratings', 'num_reviews', 'average_rating', 'rating_distribution', 'author', 'format']].copy()

# 2. Crear la tabla de Autores
authors_df = df[['author', 'authorlink']].drop_duplicates().reset_index(drop=True)
authors_df['author_id'] = authors_df.index + 1

# Vincular autores con libros en la tabla Books
books_df = pd.merge(books_df, authors_df[['author_id', 'author']], left_on='author', right_on='author', how='left').drop(columns=['author'])

# 3. Crear la tabla de Formatos
formats_df = df[['format']].drop_duplicates().reset_index(drop=True)
formats_df['format_id'] = formats_df.index + 1

# Vincular formatos con libros en la tabla Books
books_df = pd.merge(books_df, formats_df[['format_id', 'format']], left_on='format', right_on='format', how='left').drop(columns=['format'])

# books_df ahora es tu subtabla Books lista para ser usada
```

```python
# Assuming formats_df already contains a column named 'format'
split_format = formats_df["format"].str.split(",", expand=True)

# Clean up the number of pages column
formats_df["num_pages"] = (
    split_format[0]
    .str.replace("[", "", regex=False)
    .str.replace("]", "", regex=False)
    .str.replace("'", "", regex=False)
    .str.replace("pages", "", regex=False)
    .str.strip()
)

# Clean up the book format column
formats_df["book_format"] = (
    split_format[1]
    .str.replace("[", "", regex=False)
    .str.replace("]", "", regex=False)
    .str.replace("'", "", regex=False)
    .str.strip()
)

# Drop the original 'format' column as it's no longer needed
formats_df = formats_df.drop(columns=["format"])

formats_df
```

```python
# 5. Tabla de Calificaciones
ratings_df = df[['book_id', 'rating_distribution']].copy()
ratings_df
```

```python
import ast

# Función para extraer los valores de la distribución de calificaciones
def extract_ratings(rating_dist):
    # Convertir la cadena en un diccionario
    rating_dict = ast.literal_eval(rating_dist)
    
    # Extraer los valores para cada estrella
    stars_5 = int(rating_dict.get('5', '0').replace(",", ""))
    stars_4 = int(rating_dict.get('4', '0').replace(",", ""))
    stars_3 = int(rating_dict.get('3', '0').replace(",", ""))
    stars_2 = int(rating_dict.get('2', '0').replace(",", ""))
    stars_1 = int(rating_dict.get('1', '0').replace(",", ""))
    
    return pd.Series([stars_5, stars_4, stars_3, stars_2, stars_1])

# Aplicar la función a la columna rating_distribution
ratings_df[['5_star_reviews', '4_star_reviews', '3_star_reviews', '2_star_reviews', '1_star_reviews']] = ratings_df['rating_distribution'].apply(extract_ratings)

# Mostrar las primeras filas de ratings_df para confirmar
ratings_df.head()
```

```python
# Eliminar la columna original 'rating_distribution'
ratings_df = ratings_df.drop(columns=['rating_distribution'])

# Mostrar las primeras filas para confirmar que la columna se eliminó
ratings_df.head()
```

```python
# Guardar la tabla Books en un archivo CSV
books_df.to_csv('Books.csv', index=False)

# Guardar la tabla Authors en un archivo CSV
authors_df.to_csv('Authors.csv', index=False)

# Guardar la tabla Formats en un archivo CSV
formats_df.to_csv('Formats.csv', index=False)

# Guardar la tabla Genres en un archivo CSV
genres_df.to_csv('Genres.csv', index=False)

# Guardar la tabla Book_Genres en un archivo CSV
book_genres_df.to_csv('Book_Genres.csv', index=False)

# Guardar la tabla Ratings en un archivo CSV
ratings_df.to_csv('Ratings.csv', index=False)
```
